```python
import regex as re

with open('communities_names.txt', 'r') as f:
    context = f.read()
```

```python
pattern = re.compile('@attribute(.+ )')
```

```python
column_names = [name.strip() for name in pattern.findall(context)]
column_names
```

```python
df = pd.read_csv('communities_data.csv', header=None, names=column_names)
```
